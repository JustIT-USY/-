```python
import pandas as pd
import numpy as np
def three_sigma(ser1): 
    mean_value = ser1.mean()
    std_value = ser1.std()
    rule = (mean_value - 3 * std_value > ser1) | (ser1.mean() + 3 * ser1.std() < ser1)
    index = np.arange(ser1.shape[0])[rule]
    outrange = ser1.iloc[index]
    return outrange
```
